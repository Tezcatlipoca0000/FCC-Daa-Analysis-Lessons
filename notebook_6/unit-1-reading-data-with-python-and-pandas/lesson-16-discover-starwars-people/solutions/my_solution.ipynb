```python
import pandas as pd
import requests
```

```python
res = requests.get('https://swapi.dev/api/people/')
res.text
```

```python
json_res = res.json()
json_res
```

```python
df = pd.DataFrame.from_dict(json_res)
df.head()
```

```python
unpacked = pd.json_normalize(json_res)
unpacked
```

```python
df2 = pd.DataFrame(unpacked)
df2
```

```python
unpacked_results = pd.json_normalize(json_res['results'])
unpacked_results
```

```python
df.head()
```

```python
df.drop(labels='results', axis=1, inplace=True)
```

```python
unpck_df = pd.DataFrame(unpacked_results)
unpck_df.head()
```

```python
frames = [df, unpck_df]
result = pd.concat(frames, axis=1)
result.head()
```

```python
blue_eyes = result[result['eye_color'] == 'blue']
blue_eyes.head()
```
